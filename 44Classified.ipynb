```python
from sage.graphs.graph_input import from_adjacency_matrix
from sage.graphs.connectivity import connected_components
mat44=load("44mat7.sobj")
mat44=load('List44.sobj')
print("44=",len(mat44))
matDict={}
for mat in mat44 :
    mex=matrix(7,7, lambda i,j: 1 if ((mat[i,j]!=0) | (i==j)) else 0 )
    #print(mex)
    G=Graph()
    from_adjacency_matrix(G,mex-identity_matrix(7))
    r=G.connected_components()
    matType=tuple(vector([len(k) for k in r]))
    #matType.set_immutable()
    print (matType)
    if matType in matDict :
        matDict[matType]+=[mat]
    else:
         matDict[matType]=[mat]   
    print (mat)
    print("------------------")
```
