```python
def getMatType(Mat):
    vals3=[]
    for i in range(Mat.nrows()-2):
        for j in range(i+1,Mat.nrows()):
            for k in range(j+1,Mat.nrows()):
                rSum = 0
                for l in range(Mat.ncols()):
                    rSum = rSum + abs(Mat[i][l]*Mat[j][l]*Mat[k][l])
                vals3.append(abs(rSum))
    vals2=[]
    for i in range(Mat.nrows()-1):
        for j in range(i+1,Mat.nrows()):
            rSum = 0
            for l in range(Mat.ncols()):
                rSum = rSum + abs(Mat[i][l]*Mat[j][l])
            vals2.append(rSum)
    return(sorted(vals2),sorted(vals3))
            
            


MaxV=5
codeBase = MaxV+1
nR=6
cB=vector([codeBase^(nR-j) for j in range(nR+1)])
print(cB)
print()
resLen=len(res)
absres=[matrix(7,7,[sorted([abs(res[j][0][l][k]) for k in range(7)]) for l in range(7)]).transpose() for j in range(100)]
ares=[matrix(7,7,[[abs(res[j][0][l][k]) for k in range(7)] for l in range(7)]) for j in range(resLen)]
#print(cB*ares[0]),sorted(cB*ares[0].transpose()))
print(res[0][0])
print(ares[0])  
typeS=[]
Mats=[]
print(resLen)
for Min in range(resLen):
    #nType = sorted(cB*ares[Min])#sorted(cB*ares[Min].transpose())
    #nType2 = nType # sorted(cB*ares[Min].transpose()),sorted(cB*ares[Min])
    nType=getMatType(res[Min][0].transpose())
    nType2=getMatType(res[Min][0])
    if (not(nType in typeS) and not(nType2 in typeS)):
        typeS.append(nType)
        Mats.append(res[Min][0])
print(len(typeS))
for i in range(len(Mats)):
    print("------------")
    print(Mats[i])
```
